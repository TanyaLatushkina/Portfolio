```python
num_pipeline = Pipeline(steps=[
    ('knn', KNNImputer(n_neighbors=5, weights='distance')),
    ('scale',StandardScaler())
])

cat_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('one-hot',OneHotEncoder(drop='first', handle_unknown='error', sparse_output=False))
])

col_trans = ColumnTransformer(transformers=[
    ('num_pipeline', num_pipeline, num_cols),
    ('cat_pipeline', cat_pipeline, cat_cols)
    ],
    remainder='passthrough',
    n_jobs=-1)
```

### LogisticRegression

```python
%%time
def objective(trial):    
    # –°–ø–∏—Å–æ–∫ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –ø–µ—Ä–µ–±–æ—Ä–∞
    param = {
        'solver': trial.suggest_categorical('solveg', ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']),
        'C': trial.suggest_float('c', 0.001, 0.01),
        'max_iter': trial.suggest_int('max_iter', 500, 1000)
    }
                                
    # –û–ø—Ä–µ–¥–µ–ª—è—é –º–æ–¥–µ–ª—å –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è, –∫–æ—Ç–æ—Ä–æ–π –ø–µ—Ä–µ–¥–∞—é—Ç—Å—è –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã
    estimator = LogisticRegression(**param, 
                                   class_weight='balanced', 
                                   n_jobs=-1,  
                                   random_state=RANDOM_STATE)

     # –î–æ–±–∞–≤–∏–º –ø–∞–π–ø–ª–∞–π–Ω
    clf_pipeline = Pipeline(steps=[
        ('col_trans', col_trans),
        ('model', estimator)
    ])
    clf_pipeline.fit(X_train, y_train)
    
    # –í—ã—á–∏—Å–ª—è–µ–º f1_micro
    
    return cross_val_score(clf_pipeline, X_train, y_train, cv=3, scoring='f1_micro').mean()

# –ü–æ–¥–±–∏—Ä–∞–µ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
sampler = TPESampler(consider_prior=True, multivariate=True, group=True)
optuna_logreg = optuna.create_study(study_name='LogReg', direction='maximize', sampler=sampler)
# –ó–∞–ø—É—Å–∫–∞–µ–º –ø—Ä–æ—Ü–µ—Å—Å –ø–æ–¥–±–æ—Ä–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
optuna_logreg.optimize(objective, n_trials=5, n_jobs=-1, show_progress_bar=True)
```

```python
# –∑–∞–ø–∏—à–µ–º –º–µ—Ç—Ä–∏–∫—É f1 
f1_micro_logreg = round(optuna_logreg.best_value, 2)
# –ª—É—á—à–∞—è –º–µ—Ç—Ä–∏–∫–∞ –∏ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
print('F1_micro: ', round(optuna_logreg.best_value, 2))
print('best_params: ', optuna_logreg.best_params)
```

### RandomForestClassifier

```python
%%time
def objective(trial):
    # –°–ø–∏—Å–æ–∫ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –ø–µ—Ä–µ–±–æ—Ä–∞
    param = {
        'n_estimators': trial.suggest_int('n_estimators', 10, 1000),
        'criterion': trial.suggest_categorical('criterion', ['gini', 'entropy']),
        'max_depth': trial.suggest_int('max_depth', 2, 15),
        'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2', None]),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 15),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10),
    }
    
    # –û–ø—Ä–µ–¥–µ–ª—è—é –º–æ–¥–µ–ª—å –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è, –∫–æ—Ç–æ—Ä–æ–π –ø–µ—Ä–µ–¥–∞—é—Ç—Å—è –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã
    estimator = RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1, **param)
    
    # –î–æ–±–∞–≤–∏–º –ø–∞–π–ø–ª–∞–π–Ω
    clf_pipeline = Pipeline(steps=[
        ('col_trans', col_trans),
        ('model', estimator)
    ])
    clf_pipeline.fit(X_train, y_train)
    
    # –í—ã—á–∏—Å–ª—è–µ–º f1_micro
    
    return cross_val_score(clf_pipeline, X_train, y_train, cv=3, scoring= 'f1_micro').mean()

# –ü–æ–¥–±–∏—Ä–∞–µ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
sampler = TPESampler(consider_prior=True, multivariate=True, group=True)
optuna_rf = optuna.create_study(study_name='RF', direction="maximize", sampler=sampler)
# –ó–∞–ø—É—Å–∫–∞–µ–º –ø—Ä–æ—Ü–µ—Å—Å –ø–æ–¥–±–æ—Ä–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
optuna_rf.optimize(objective, n_trials=5, n_jobs=-1, show_progress_bar=True)
```

```python
# –∑–∞–ø–∏—à–µ–º –º–µ—Ç—Ä–∏–∫—É f1 
f1_micro_rf = round(optuna_rf.best_value, 2) # –∑–∞–ø–∏—à–µ–º –º–µ—Ç—Ä–∏–∫—É f1
# –ª—É—á—à–∞—è –º–µ—Ç—Ä–∏–∫–∞ –∏ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
print('F1_micro: ', round(optuna_rf.best_value, 2))
print('best_params: ', optuna_rf.best_params)
```

### CatBoostClassifier

```python
%%time
def objective(trial):    
    # –°–ø–∏—Å–æ–∫ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –ø–µ—Ä–µ–±–æ—Ä–∞
    param = {
        'depth': trial.suggest_int('depth', 1, 12),
        'boosting_type': trial.suggest_categorical('boosting_type', ['Ordered', 'Plain']),
        'bootstrap_type': trial.suggest_categorical('bootstrap_type', ['Bayesian', 'Bernoulli', 'MVS']),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.2),
        'l2_leaf_reg': trial.suggest_int('l2_leaf_reg', 1, 7),
        'iterations': trial.suggest_int('iterations', 100, 500),
        'border_count': trial.suggest_int('border_count', 5, 20),
        'thread_count': trial.suggest_int('thread_count', 3, 4),
        'early_stopping_rounds': trial.suggest_int('early_stopping_rounds', 9, 10),
        'random_strength': trial.suggest_float('random_strength', 0.01, 1),
        'max_ctr_complexity': trial.suggest_int('max_ctr_complexity', 1, 3)
    }
                                
    # –û–ø—Ä–µ–¥–µ–ª—è—é –º–æ–¥–µ–ª—å –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è, –∫–æ—Ç–æ—Ä–æ–π –ø–µ—Ä–µ–¥–∞—é—Ç—Å—è –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã
    estimator = CatBoostClassifier(**param, verbose=False, loss_function='MultiClass')

    # –î–æ–±–∞–≤–ª—è–µ–º –ø–∞–π–ø–ª–∞–π–Ω—ã
    clf_pipeline = Pipeline(steps=[
            ('col_trans', col_trans),
            ('model', estimator)
        ]) 
    clf_pipeline.fit(X_train, y_train)
    
    # –í—ã—á–∏—Å–ª—è–µ–º f1_micro
    return cross_val_score(clf_pipeline, X_train, y_train, cv=3, scoring= 'f1_micro').mean()

# –ü–æ–¥–±–∏—Ä–∞–µ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
sampler = TPESampler(consider_prior=True, multivariate=True, group=True)
optuna_cbc = optuna.create_study(study_name='CBC', direction="maximize", sampler=sampler)
# –ó–∞–ø—É—Å–∫–∞–µ–º –ø—Ä–æ—Ü–µ—Å—Å –ø–æ–¥–±–æ—Ä–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
optuna_cbc.optimize(objective, n_trials=5, n_jobs=-1, show_progress_bar=True)
```

```python
# –∑–∞–ø–∏—à–µ–º –º–µ—Ç—Ä–∏–∫—É f1 
f1_micro_cbc = round(optuna_cbc.best_value, 2)
# –ª—É—á—à–∞—è –º–µ—Ç—Ä–∏–∫–∞ –∏ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
print('F1_micro: ', round(optuna_cbc.best_value, 2))
print('best_params: ', optuna_cbc.best_params)
```
