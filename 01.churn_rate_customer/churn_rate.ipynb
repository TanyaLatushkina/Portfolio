## Проведем предобработку данных

### Соединим датасеты

```python
# соединим датафреймы
df = contract.merge(personal, on='customer_id', how='inner').\
              merge(internet, on='customer_id', how='outer').\
              merge(phone, on='customer_id', how='outer')
```

```python
# основная информация
data_info(df)
```

```python
# сделаем копию датафрейма
df_temp = df.copy()
```

### Проведем предобработку данных

#### Предобработаем данные

__senior_citizen__

```python
df['senior_citizen'] = df['senior_citizen'].astype('object')
```

__total_charges__

```python
# total_charges изучим пропущенные значения
df[df['total_charges'].isna()]
```

  Нулевые значения в признаке total_charges из-за того, что договор заключался 01.02.2020. Это последняя дата наблюдений за пользователями и платежи просто не успели пройти. Заменим пропуска на 0.
  
  На основании этой информации, можно предположить, что договора оплачиваются не авансом, а фактическим платежом.

```python
df['total_charges'] = df['total_charges'].fillna(0)
```

__Заполним нулевые значения, получившиеся в итоге слияния датафреймов__

```python
# запишем в переменную все признаки, которые относят к интернет услугам
internet_services = ['internet_service',\
                     'online_security',
                     'online_backup',
                     'device_protection',
                     'tech_support',
                     'streaming_tv',
                     'streaming_movies']
```

```python
# заменим нулевые значения
df[internet_services] = df[internet_services].fillna('witout_internet')
```

```python
# заменим нулевые значения
df['multiple_lines'] = df['multiple_lines'].fillna('witout_phone')
```

__Выделим таргет__

```python
# уникальные даты окончания договоров
df['end_date'].unique()
```

```python
# сгенирим целевой признак
df['target'] = df['end_date'].apply(lambda x: 0 if x==None else 1)
```

```python
# посмотри распределение
df['target'].value_counts()
```

__Заполним пропуска в признаке end_date__

```python
#Так как мы точно знаем, что на мометь 1 февряля 2020 года договор не был разорван, заполним пропуска следующей датой
df['end_date'] = pd.to_datetime(df['end_date']) # поменяем тип данных
df['end_date'] = df['end_date'].fillna(pd.to_datetime('2020-02-02', format='%Y-%m-%d')) # заполним пропуска
```

__Создадим признак длительности договора и удалим временные признаки__

```python
def before_payment(row):

    ''' 
        Функция подсчета дней до очередного платежа по договору
        На вход получает нужные признак, на выходе новый признак сколько осталось дней до очередного платежа
    '''

    if row['type'] == 'Month-to-month':
        return round(row['duration_contract'] % 30)
    elif row['type'] == 'One year':
        return round(row['duration_contract'] % 365)
    elif row['type'] == 'Two year':
        return round(row['duration_contract'] % 730)
```

```python
# создадим pipeline для обработки
num_pipeline = Pipeline(steps=[
    ('scale', StandardScaler())
])

cat_pipeline = Pipeline(steps=[
    ('encoder', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=7))
])
```

```python
col_trans_cb = ColumnTransformer(transformers=[   # pipeline  для RandomForesClassifier
    ('num_pipeline', num_pipeline, num_features),
    ],
    remainder='passthrough',
    n_jobs=-1)

col_trans_rf = ColumnTransformer(transformers=[   # pipeline для CatBoostClassifier
    ('num_pipeline', num_pipeline, num_features),
    ('cat_pipeline', cat_pipeline, cat_features)
    ],
    remainder='passthrough',
    n_jobs=-1)
```

```python
def get_col_name(pipe) -> list:

    ''' 
        Функция получение нормальные наименований признаков после работы pipeline.
        На вход получаетя pipeline трансформер признаков
        На выходе нормальное название признаков после работы pipeline
    '''
    column_name = pd.DataFrame(pipe.get_feature_names_out())
    column_name = column_name[0].apply(lambda x: x.split('__'))
    column_name = column_name.apply(lambda x: x[1]).tolist()
    return column_name
```

```python
# Подготовлю данные для моделей.
X_train_rf = pd.DataFrame(col_trans_rf.fit_transform(X_train)) # признаки для RandomForestClassifier
X_train_rf.columns = get_col_name(col_trans_rf)                # вытаскиваем нормально название признаков

X_train_cb = pd.DataFrame(col_trans_cb.fit_transform(X_train)) # признаки для CatBoostClassifier
X_train_cb.columns = get_col_name(col_trans_cb)                # вытаскиваем нормально название признаков
```

## Обучаем модели

### RandomForestClassifier

```python
# подбор параметров и обучение модели RandomForestClassifier

def objective(trial):
    # Список гиперпараметров для перебора

    param = {
        'n_estimators': trial.suggest_int('n_estimators', 10, 1000),
        'max_depth': trial.suggest_int('max_depth', 2, 15),
        'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2', None]),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 15),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10),
    }

    # Определяю модель машинного обучения, которой передаются гиперпараметры

    estimator = RandomForestClassifier(**param, random_state=SEED, n_jobs=-1) #, class_weight= 'balanced')

    # Обучим модель

    estimator.fit(X_train_rf, y_train)

    # Вычисляем roc_auc

    return cross_val_score(estimator, X_train_rf, y_train, cv=5, scoring= 'roc_auc').mean()
```

```python
# Подбираем гиперпараметров

sampler = TPESampler(consider_prior=True, multivariate=True, group=True, seed=SEED)
optuna_rf = optuna.create_study(study_name='RF', direction='maximize', sampler=sampler)
```

```python
# Запускаем процесс подбора гиперпараметров
optuna_rf.optimize(objective, n_trials=5, n_jobs=-1, show_progress_bar=True)
```

```python
# лучшая метрика и параметры
print('Roc_auc: ', round(optuna_rf.best_value, 2))
print('best_params: ', optuna_rf.best_params)
```

### CatBoostClassifier

```python
# подбор параметров и обучение модели CatBoostClassifier

def objective(trial):
    # Список гиперпараметров для перебора
    param = {
        'depth': trial.suggest_int('depth', 1, 12),
        'boosting_type': trial.suggest_categorical('boosting_type', ['Ordered', 'Plain']),
        'bootstrap_type': trial.suggest_categorical('bootstrap_type', ['Bayesian', 'Bernoulli', 'MVS']),
        'learning_rate': trial.suggest_float('learning_rate', 0.0001, 0.2),
        'l2_leaf_reg': trial.suggest_int('l2_leaf_reg', 1, 7),
        'iterations': trial.suggest_int('iterations', 100, 500),
        'border_count': trial.suggest_int('border_count', 5, 20),
        'thread_count': trial.suggest_int('thread_count', 3, 4),
        'early_stopping_rounds': trial.suggest_int('early_stopping_rounds', 9, 10),
        'random_strength': trial.suggest_float('random_strength', 0.01, 1),
        'max_ctr_complexity': trial.suggest_int('max_ctr_complexity', 1, 3)
    }


    estimator = CatBoostClassifier(**param,
                                   verbose=False,
                                   cat_features=cat_features,
                                   loss_function='Logloss',
                                   random_state=SEED)

    # Обучим модель

    estimator.fit(X_train_cb, y_train)

    # Вычисляем roc_auc

    return cross_val_score(estimator, X_train_cb, y_train, cv=5, scoring='roc_auc').mean()
```

```python
# Подбираем гиперпараметров

sampler = TPESampler(consider_prior=True, multivariate=True, group=True, seed=SEED)
optuna_cbc = optuna.create_study(study_name='CBC', direction='maximize', sampler=sampler)
# Запускаем процесс подбора гиперпараметров

optuna_cbc.optimize(objective, n_trials=5, n_jobs=-1, show_progress_bar=True)
```

```python
# лучшая метрика и параметры
print('Roc_auc: ', round(optuna_cbc.best_value, 2))
print('best_params: ', optuna_cbc.best_params)
```

График Roc_auc отображает соотношение между долей объектов от общего количества носителей признака, верно классифицированных как несущие признак, и долей объектов от общего количества объектов, не несущих признака, ошибочно классифицированных как несущие признак. Чем бошьше площадь под кривой, тем лучше качество классификации. Тут мы видим неплохо результат относительно  коэффиценки 0.5 (случайные угадывания).

```python
# посчитаем precision и recall
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_proba)

#создадим граффик зависимости precision и recall
fig, ax = plt.subplots()
ax.plot(recall, precision, color='purple')

#добавим подписи
ax.set_title('Precision-Recall Curve')
ax.set_ylabel('Precision')
ax.set_xlabel('Recall')
plt.show()
```

Кривая Precision-Recall — это график со значениями точности на оси Y и значениями отзыва на оси X.
Чем больше площадь под кривой, тем лучше работает модель. Площадь под кривой в нашем случае равна 0.79.
